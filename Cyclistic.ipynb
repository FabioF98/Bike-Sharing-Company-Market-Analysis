## About Cyclistic
Cyclistic, a Chicago Department of Transportation program, is a bike-share system in Chicago and Evanston. Cyclistic is a convenient, fun, and affordable transportation option for commuting to work, getting around town, and exploring Chicago. Like other bike-share systems, Cyclistic consists of a fleet of specially designed, geotracked, and durable bikes that are locked into a network of docking stations throughout the region. The bikes can be unlocked from one station and returned to any other station in the system 24/7.

## How can Cyclistic drive more annual memberships?
Cyclistic marketing relies on building general awareness and a well-defined set of value propositions that appeal to broad consumer segments. This approach led to great market awareness for the bike-share program and diverse marketing tactics that promoted its many benefits. One such benefit was flexibility of its pricing plans. Consumers could use a bike for short trips by buying a single-ride pass or for an entire day through a full-day pass (internally, Cyclistic referred to the consumers as "casual riders"). Alternatively consumers could become annual members to gain unlimited access to Cyclistic bikes for the entire year (customers Cyclistic called "members"). Cyclistic finance analysts concluded that members were much more profitable than casual riders (both single ride and full day).<br/><br>
Moreno knew the easiest path to annual memberships was to convince Chicago-area casual riders to become members. These customers were already aware of the Cyclistic program, which made messaging easier, as there was no need to cover the basics of the Cyclistic experience. Furthermore, casual riders had chosen Cyclistic for their mobility needs, demonstrating consideration for the program and making them easier to guide toward becoming members than consumers who might want to first try the bike-share program before committing to the annual plan.<br/><br>
Data privacy introduced one important complication. The city of Chicago’s recent regulatory changes and data privacy commitments prohibited Moreno and her team from working with casual riders’ personally identifiable information. In other words, Cyclistic was precluded from connecting pass purchases to individuals. Linking credit card numbers to passes could help the Cyclistic marketing team segment casual riders by their home addresses (i.e., separating those who live in the Cyclistic service area from tourists), consumers who bought multiple or repeat passes, and other important views into the data.<br/><br>
Moreno turned to her team of data analysts for help, she was sure that answers to her questions could be found in the Cyclistic trip data. The Cyclistic executive team would require data-rooted answers to the questions Moreno handed her analytics team.<br/><br><br/><br>

## 4\) Analyze Data
* Removed columns for geographic data as it's no longer in use
* Removed data relative to bike maintenance or quality controls, this type of data was marked using all caps in their station names. All the rows having this data were removed since some of them contained long periods of time where bikes were taken out of service for maintenance and their data would have influenced the average ride duration. Later in the analysis more data having ride duration which were hard to justify was found, with some being as long as a month. Every ride with an extremely long duration time was done by customers using docked bikes, which led me to think it was a system error, data including the excessively long rides was removed to avoid compromising the analysis. 
* Serious inconsistencies in station IDs were found, IDs were not unique to a station and they also changed from year 2020 to year 2021, moreover the composition of IDs was impossible to establish, as some of them contained only letters, other only numbers, others contained both. 
* Removed rows containing negative ride duration (probably due to system errors) and rows containing ride duration of less than 60 seconds, as they did not represent real usage of the service, but realistically meant a service activated by mistake and then immediately terminated or users checking whether they correctly terminated the previous ride.
* Descriptive analysis showed a significance difference in usage time between customers and annual subscribers, with the former having longer rides on average. Interesting trends were visible once the visualizations were created. Daily trends showed that customers use the service less than subscribers, even though peak time of usage in both cases was at 18, it's worth mentioning that subscribers use the service more at times when they need to commute to and from work or university, meaning that some of them could use the service as their main transportation during the day (not enough data to support this finding). Weekly trends showed that customers use the service the most on weekdays with a peak on Saturdays and Sundays, whereas Subscribers use the service constantly throughout the week (using it at most on Wednesdays) with a slight drop on weekends. Monthly trends showed that peak usage in customers was during months when tourism is at its highest and generally during the summer. There's a severe drop in usage during winter, with the number of customers extremely reduced, subscribers still tend to use the service during winter months but less with respect to their average use during the rest of the year. It's interesting to notice that users with an annual subscription tend to use the service more despite the harsh climate conditions of winter months, there's still a significant drop in usage but overall subscribers suffer less from the drops in usage of winter time.

```r
info <- all_trips_v6  %>%
group_by(user_type) %>%
summarise(number_of_users = n())  %>% 
mutate(percent = percent(round(number_of_users/sum(number_of_users), 3)))
info
```

```r
options(repr.plot.width = 10, repr.plot.height = 7)
ggplot(info, aes(x = "", y = number_of_users, fill = user_type)) +
  geom_bar(width = 1, stat = "identity", color ="black") +
  geom_text(aes(label = paste(round(number_of_users / sum(number_of_users) * 100, 1), "%")),
    color = "#ffffff", size = 10,
    position = position_stack(vjust = 0.5)) +
  scale_fill_manual(name = "User Type", values = c("#bb8b61", "#3a5f73")) +
  coord_polar("y") +
  xlab("") + ylab("") +
theme(axis.text = element_blank(),
  axis.ticks = element_blank(),
  plot.title = element_text(hjust = 0.5, lineheight = 2, size = 20),
  legend.key.size = unit(2.5, "line"),
  legend.text = element_text(size = 14),
  legend.title = element_text(size = 18)) +
labs(title = "Distribution of Users")
```

I checked the bike preference for each user type in order to understand what a future with more subscribers would look like. Getting more customers to subscribe to an annual plan would imply a reduction in usage of docked bikes (although it's worth noticing that data on docked bikes could be inaccurate as we've seen before) and an increase in usage of classic bikes. This, combined with information relative to their preferred bike stations, could help allocate more bikes where they are needed.

```r
options(repr.plot.width = 14, repr.plot.height = 7)
customer_data <- all_trips_v6  %>%
group_by(bike_type, user_type) %>%
summarise(number_of_users = n(), .groups = "drop")  %>%
filter(user_type == "Customer")  %>% 
mutate(percent = percent(number_of_users/sum(number_of_users)))

subscriber_data <-  all_trips_v6  %>%
group_by(bike_type, user_type) %>%
summarise(number_of_users = n(), .groups = "drop")  %>%
filter(user_type == "Subscriber")  %>% 
mutate(percent = percent(number_of_users / sum(number_of_users)))
customer_data
subscriber_data
```

```r
options(repr.plot.width = 16, repr.plot.height = 7)
customer_pie <- ggplot(data = customer_data, aes (x = "", y = number_of_users,
  fill = bike_type)) +
geom_bar(width = 1, stat = "identity", color = "black") +
geom_text(aes(label = paste(round(number_of_users / sum(number_of_users)* 100), "%")),
  color = "#ffffff", size = 10,
  position = position_stack(vjust = 0.5)) +
scale_fill_manual(name = "User Type",
  values = c("#3a5f73", "#ecc19c", "#a1695b"),
  guide = "none") +
coord_polar("y") +
xlab("") + ylab("") +
theme (axis.text = element_blank(),
  axis.ticks = element_blank(),
  plot.title = element_text(hjust = 0.5, lineheight = 2, size = 20)) +
labs(title = "Bike Preference for Customers")

subscriber_pie <- ggplot(data = subscriber_data, aes (x = "", y = number_of_users,
  fill = bike_type)) +
geom_bar(width = 1, stat = "identity", color = "black") +
  geom_text(aes(label = paste(round(number_of_users / sum(number_of_users)* 100), "%")),
  color = "#ffffff", size = 10,
  position = position_stack(vjust = 0.5)) +
scale_fill_manual(name = "User Type",
  labels = c("Classic Bike", "Docked Bike", "Electric Bike"),
  values = c("#3a5f73","#ecc19c", "#a1695b")) +
coord_polar("y") +
xlab("") + ylab("") +
theme(axis.text = element_blank(),
  axis.ticks = element_blank(),
  plot.title = element_text(hjust = 0.5, lineheight = 2, size = 20),
  legend.key.size = unit(2.5, "line"),
  legend.text = element_text(size = 14),
  legend.title = element_text(size = 18)) +
labs(title = "Bike Preference for Subscribers")

plot_grid(customer_pie, subscriber_pie, align = "hv", axis = "tblr")
```

I then moved on to the monthly usage, I first created a table to get a quantitative summary of the data, then I created a visualization that showed annual trends

```r
options(repr.plot.width = 10, repr.plot.height = 7)
all_trips_v6  %>%
mutate(weekday = wday(start_time, label = TRUE))  %>%
group_by(user_type, month)  %>%
summarise(number_of_rides = n(), .groups = "drop") %>%
arrange(month)  %>%
rename("User Type" = user_type, "Month" = month,
  "Number of Rides" = number_of_rides)
```

```r
options(repr.plot.width = 14, repr.plot.height = 7)
all_trips_v6  %>%
group_by(user_type, month) %>%
summarise(number_of_users = n(), .groups = "drop") %>%
arrange(month)  %>%
ggplot(aes(x = month, y = number_of_users, color = user_type, group = user_type)) +
scale_color_manual(name = "User Type", values = c("#e4ad7d", "#3a5f73")) +
geom_point(stat = "identity", size = 2) + geom_line(size = 1) +
scale_y_continuous(labels = scales::comma) +
theme(plot.title = element_text(hjust = 0.5, lineheight = 2, size = 20),
  plot.subtitle = element_text(hjust = 0.5, size = 15),
  axis.title.x = element_blank(),
  axis.title.y = element_blank(),
  legend.key.size = unit(2.5, "line"),
  legend.text = element_text(size = 14),
  legend.title = element_text(size = 18)) +
labs(title = "Number of Users per Month",
  subtitle = "Customers vs Subscribers",
  fill = "User Type")
```

The graph above showed that seasonal changes affect the usage of bikes in both customers and subscribers. In particular we can see an important drop in both numbers during the winter time (to be expected, since harsh climate conditions don't make for a comfortable bike ride). One interesting observation is that people who have a subscription tend to use the service more throughout the year, even in the coldest months. Usage of customers is extremely low in these months. 

Another feature that I wanted to analyze was the average ride duration of each type of user during the week

```r
all_trips_v6  %>%
mutate(weekday = wday(start_time, label = TRUE))  %>%
group_by(user_type, day_of_week)  %>% 
summarise(average_duration = mean(ride_duration), .groups = "drop")  %>%
rename("User Type" = user_type, "Day of the Week" = day_of_week,
  "Average Duration" = average_duration)
```

The table showed that customers tend to use the bikes for longer period of times, this could be due to customers using the daily pass (no data to support this hypothesis), customers have to choose between two type of passes, single-ride and daily, while subscribers are free to take and drop bikes whenever they want since they are charged for the annual membership and not for single rides.

Once I saw the weekly trends I created a new table comparing the overall mean, median, maximum and minimum ride duration between customers and subscribers.

```r
setNames(aggregate(all_trips_v6$ride_duration ~ all_trips_v6$user_type, FUN = mean), c("User Type", "Ride Duration"))
setNames(aggregate(all_trips_v6$ride_duration ~ all_trips_v6$user_type, FUN = median), c("User Type", "Ride Duration"))
setNames(aggregate(all_trips_v6$ride_duration ~ all_trips_v6$user_type, FUN = max), c("User Type", "Ride Duration"))
setNames(aggregate(all_trips_v6$ride_duration ~ all_trips_v6$user_type, FUN = min), c("User Type", "Ride Duration"))
```

```r
options(repr.plot.width = 13, repr.plot.height = 9)
ggplot(data = all_trips_v6) +
geom_bar(aes(x = day_of_week, fill = bike_type), color = "#666666") +
facet_wrap(~user_type) + 
scale_fill_manual(name = "Bike typology",
labels = c("Classic Bike", "Docked Bike", "Electric Bike"),
  values = c("#3a5f73", "#ecc19c", "#a1695b")) +
labs(title = "Weekly Usage Comparison",
  subtitle = "(Customer vs. Subscriber)") +
scale_y_continuous(labels = scales::comma) +
theme_bw(base_size = 15) +
theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
ylab("") + xlab("")
```

Data shows that peak usage of bikes for casual members occurs during the weekend, this could imply that mainly tourists or occasional riders use the service in this time frame. Data also shows that annual members don't use docked bikes as much as casual members and that their overall usage of the bike service during the week is more constant. The trend for annual members is opposite to those of casual members, as you can see the peak usage is on Wednesdays and tends to be steady throughout the week with a slight drop on Sundays.

I checked what were the ten most common stations. As expected, the top stations for customers contain several tourist attractions and are different from the ones that subscribers use more often. This tells us that a good part of customers could be tourists, it would be useful to distinguish between tourists and users that don't have an annual subscription but still live in the service area of Cyclistic to better target the audience for the advertisement.
